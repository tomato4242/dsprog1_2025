```python
import requests
from bs4 import BeautifulSoup
import time
from urllib.parse import urljoin, urlparse

surl = "https://www.musashino-u.ac.jp"
DOMAIN = urlparse(surl).netloc

to_crawl = [surl]
crawled = set()
mapsisemap = {}
count = 0

print("クロールを始める！！！\n")

while to_crawl:
    wwith_crawl = to_crawl.pop(0)

    if wwith_crawl in crawled:
        continue

    crawled.add(wwith_crawl)

    try:
        time.sleep(1.11)
        
        res = requests.get(wwith_crawl)
        res.encoding = res.apparent_encoding 
        
        oisisounasup = BeautifulSoup(res.text, "html.parser")

        t_tag = oisisounasup.find("title")
        title = t_tag.string.strip() if t_tag and t_tag.string else "（タイトルなし）"
        
        mapsisemap[wwith_crawl] = title
        
        count += 1
        print(f"[{count}] {title}\n  -> {wwith_crawl}")

        for link in oisisounasup.find_all('a', href=True):
            href = link.get('href') 

            absolute_url = urljoin(wwith_crawl, href)
            
            parsed_url = urlparse(absolute_url)
            normalized_url = parsed_url._replace(query="", fragment="").geturl()

            if (urlparse(normalized_url).netloc == DOMAIN and
                normalized_url not in crawled and      
                normalized_url not in to_crawl and        
                not any(normalized_url.lower().endswith(ext) for ext in ['.pdf', '.jpg', '.jpeg', '.png', '.gif', '.svg','bmg','.ico','webp',])): 
                
                to_crawl.append(normalized_url)

    except Exception as e:
        print(f"  -> エラーが発生しました: {e}")

# --- 変更箇所はここからです ---

print("\nクロールが完了しました！！！")
print("\n--- サイトマップ ---")
for url, title in mapsisemap.items():
    print(f"・タイトル：{title}")
    print(f"  URL：{url}")
    print("---") 

print(f"\n合計 {len(mapsisemap)} ページを検出しました。")
```
